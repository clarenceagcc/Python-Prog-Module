```python
import pandas as pd
import numpy as np
import ast
```

```python
df = pd.read_csv('out_with_pros.csv')   #import file
df.head() # Getting basic information
```

```python
# information about columns and data types
# no null values
df.info()
```

```python
# copying dataset for cleaning
df_clear = df.copy()
```

```python
# Getting names of columns
df.columns
```

```python
df_clear.rename(columns=str.lower, inplace=True)
df_clear
```

```python
df_clear.columns = df_clear.columns.str.replace(' ', '_')
df_clear.columns = df_clear.columns.str.replace('&', '_')
df_clear
```

```python
company_name = df_clear.company.tolist()
# removed the numbers from the company name
for i in range(len(company_name)):
    company_name[i] = company_name[i].split('\n',1)[0]
```

```python
# preview cleaned company names
df_clear.company = company_name
df_clear.head()
```

```python
# Replacing -1 values to 0-0 for easier processing of max/min salary values
salary_estimate = df_clear.salary.tolist()

for i in range(len(salary_estimate)):
    if salary_estimate[i] == '-1':
        salary_estimate[i] = '$0K-$0K'
```

```python
# Extract salary value by texts' (Employer Provided Salary: and  (Glassdoor Est.))
for i in range(len(salary_estimate)):
    salary_estimate[i] = salary_estimate[i].replace('Employer Provided Salary:', '')
    salary_estimate[i] = salary_estimate[i].replace(' (Glassdoor Est.)', '')
    salary_estimate[i] = salary_estimate[i].replace('$', '')
    salary_estimate[i] = salary_estimate[i].replace('K', '')
    salary_estimate[i] = salary_estimate[i].replace(' ', '')

    if 'Perhour' in salary_estimate[i]:
        salary_estimate[i] = '12-12'

print(salary_estimate)
```

```python
df_clear.salary = salary_estimate
df_clear.salary.unique()
```

```python
# Splitting data to make a low and max salary limit columns
split_data = df_clear.salary.str.split('-', expand=True)
low_salary_limit = split_data[0].tolist()
max_salary_limit = split_data[1].tolist()

for i in range(len(max_salary_limit)):
    if max_salary_limit[i] == None:
        max_salary_limit[i] = low_salary_limit[i]
```

```python
new_idx = df_clear.columns.get_loc('salary') + 1
```

```python
# insert the low_salary_limit column after the salary_estimate column
df_clear.insert(new_idx, 'low_salary_limit', low_salary_limit)

new_idx2 = df_clear.columns.get_loc('low_salary_limit') + 1
```

```python
# insert the max_salary_limit column after the low_salary_limit column
df_clear.insert(new_idx2, 'max_salary_limit', max_salary_limit)

#pd.set_option('display.max_rows', None)
#df_clear
# remove extra values from the column with salary
df_clear['max_salary_limit'] = df_clear['max_salary_limit'].astype(int)
df_clear['low_salary_limit'] = df_clear['low_salary_limit'].astype(int)
```

```python
# lead to thousands
df_clear[['low_salary_limit', 'max_salary_limit']] = df_clear[['low_salary_limit', 'max_salary_limit']] * 1000
```

```python
# remove the salary estimate column
df_clear.drop(columns='salary', inplace=True)
df_clear
```

```python
# Converting company pros and cons from str to list, since data in df is str
companypros = df_clear.company_pros_reviews.tolist()
companycons = df_clear.company_cons_reviews.tolist()

for i in range(len(companypros)):
    companypros[i] = ast.literal_eval(companypros[i])
    companycons[i] = ast.literal_eval(companycons[i])
```

```python
# Removing all -1 values from pro and con reviews (might have to run this cell at least 5 times)
for i in range(len(companypros)):
    for x in companypros[i]:
        if x == -1:
            companypros[i].remove(x)

    for x in companycons[i]:
        if x == -1:
            companycons[i].remove(x)
```

```python
for i in range(len(companypros)):
    if not companypros[i]:
        companypros[i] = 'N/A'
    if not companycons[i]:
        companycons[i] = 'N/A'

df_clear.company_pros_reviews = companypros
df_clear.company_cons_reviews = companycons
```
