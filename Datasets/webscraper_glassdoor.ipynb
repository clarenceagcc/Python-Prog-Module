```python
from selenium.common.exceptions import NoSuchElementException, ElementClickInterceptedException
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium import webdriver
import time
import pandas as pd
```

```python
def get_jobs(num_jobs, verbose):
    links = ['https://www.glassdoor.sg/Job/information-technology-jobs-SRCH_KO0,22.htm?fromAge=14',
             'https://www.glassdoor.sg/Job/programmer-jobs-SRCH_KO0,10_IP2.htm?includeNoSalaryJobs=true',
            'https://www.glassdoor.sg/Job/technical-support-specialist-jobs-SRCH_KO0,28.htm',
            'https://www.glassdoor.sg/Job/programmer-jobs-SRCH_KO0,10.htm',
            'https://www.glassdoor.sg/Job/technical-support-engineer-jobs-SRCH_KO0,26_IP2.htm?includeNoSalaryJobs=true',
            'https://www.glassdoor.sg/Job/cyber-security-jobs-SRCH_KO0,14.htm',
            'https://www.glassdoor.sg/Job/technical-support-engineer-jobs-SRCH_KO0,26.htm',
            'https://www.glassdoor.sg/Job/full-stack-developer-jobs-SRCH_KO0,20.htm',
             'https://www.glassdoor.sg/Job/software-engineer-jobs-SRCH_KO0,17_IP3.htm?includeNoSalaryJobs=true',
            'https://www.glassdoor.sg/Job/full-stack-engineer-jobs-SRCH_KO0,19.htm',
            'https://www.glassdoor.sg/Job/machine-learning-engineer-jobs-SRCH_KO0,25.htm',
            'https://www.glassdoor.sg/Job/data-scientist-jobs-SRCH_KO0,14.htm',
            'https://www.glassdoor.sg/Job/big-data-engineer-jobs-SRCH_KO0,17.htm',
             'https://www.glassdoor.sg/Job/network-administrator-jobs-SRCH_KO0,21.htm'
            'https://www.glassdoor.sg/Job/server-administrator-jobs-SRCH_KO0,20.htm',
            'https://www.glassdoor.sg/Job/front-end-engineer-jobs-SRCH_KO0,18.htm',
            'https://www.glassdoor.sg/Job/front-end-engineer-jobs-SRCH_KO0,18_IP2.htm?includeNoSalaryJobs=true',
            'https://www.glassdoor.sg/Job/back-end-developer-jobs-SRCH_KO0,18.htm',
             'https://www.glassdoor.sg/Job/back-end-developer-jobs-SRCH_KO0,18_IP2.htm?includeNoSalaryJobs=true',
             'https://www.glassdoor.sg/Job/game-developer-jobs-SRCH_KO0,14.htm'
             'https://www.glassdoor.sg/Job/game-developer-jobs-SRCH_KO0,14_IP2.htm?includeNoSalaryJobs=true',
             'https://www.glassdoor.sg/Job/ai-jobs-SRCH_KO0,2.htm',
             'https://www.glassdoor.sg/Job/ai-jobs-SRCH_KO0,2_IP2.htm?includeNoSalaryJobs=true',
            'https://www.glassdoor.sg/Job/cloud-solution-architect-jobs-SRCH_KO0,24.htm',
             'https://www.glassdoor.sg/Job/cloud-solution-architect-jobs-SRCH_KO0,24_IP2.htm?includeNoSalaryJobs=true',
            'https://www.glassdoor.sg/Job/software-engineer-jobs-SRCH_KO0,17.htm',
            'https://www.glassdoor.sg/Job/software-engineer-jobs-SRCH_KO0,17_IP2.htm?includeNoSalaryJobs=true',
            'https://www.glassdoor.sg/Job/it-security-specialist-jobs-SRCH_KO0,22.htm',
            'https://www.glassdoor.sg/Job/user-experience-designer-jobs-SRCH_KO0,24.htm',
            'https://www.glassdoor.sg/Job/web-developer-jobs-SRCH_KO0,13.htm',
            'https://www.glassdoor.sg/Job/database-developer-jobs-SRCH_KO0,18.htm',
            'https://www.glassdoor.sg/Job/quality-assurance-tester-jobs-SRCH_KO0,24.htm',
            'https://www.glassdoor.sg/Job/quality-assurance-tester-jobs-SRCH_KO0,24_IP2.htm?includeNoSalaryJobs=true',
            'https://www.glassdoor.sg/Job/web-developer-jobs-SRCH_KO0,13_IP2.htm?includeNoSalaryJobs=true',
            'https://www.glassdoor.sg/Job/it-security-specialist-jobs-SRCH_KO0,22_IP2.htm?includeNoSalaryJobs=true'
            'https://www.glassdoor.sg/Job/information-technology-jobs-SRCH_KO0,22.htm?fromAge=14',]
    
    '''Gathers jobs as a dataframe, scraped from Glassdoor'''
    
    #Initializing the webdriver
    options = webdriver.ChromeOptions()
    
    #Uncomment the line below if you'd like to scrape without a new Chrome window every time.
    #options.add_argument('headless')
    
    #Change the path to where chromedriver is in your home folder.
    service = Service(executable_path='/usr/bin/chromedriver')
    driver = webdriver.Chrome(service=service, options=options)
    jobs = []
    driver.set_window_size(1280, 1000)
    collected_all = False
    counter = 0
    
    while not collected_all:  #If true, should be still looking for new jobs.
        if len(jobs) != num_jobs:
            for pages in range(0,35):
                if pages == 33:
                    collected_all = True
                if len(jobs) == num_jobs:
                    break
                print(f'PAGES : {pages}')
                url = links[pages]
                driver.get(url)
        
                time.sleep(.1)
                job_buttons = driver.find_elements(By.CLASS_NAME, 'react-job-listing')
        
                #Test for the "Sign Up" prompt and get rid of it.
                try:
                    driver.find_element(By.CLASS_NAME, 'selected').click()
                except ElementClickInterceptedException:
                    pass
        
                time.sleep(1)
        
                try:
                    driver.find_element(By.XPATH, '/html/body/div[12]/div/div/div/div[2]/button').click()  #clicking to the X.
                except NoSuchElementException:
                    pass
                
                try:
                    driver.find_element(By.XPATH, '//*[@id="LoginModal"]/div/div/div/div[2]/button').click()
                except NoSuchElementException:
                    pass
                for jobs_ in job_buttons:
                    print("Progress: {}".format("" + str(len(jobs)) + "/" + str(num_jobs)))
    
                    if len(jobs) == num_jobs:
                        break
                    
                    time.sleep(1)
                    jobs_.click()
                    collected_successfully = False
                    while not collected_successfully:
                        try:
                            job_title = driver.find_element(By.XPATH, '//*[@id="JDCol"]/div/article/div/div[1]/div/div/div[1]/div[3]/div[1]/div[2]').text
                            company_name = driver.find_element(By.XPATH,'//*[@id="JDCol"]/div/article/div/div[1]/div/div/div[1]/div[3]/div[1]/div[1]/div').text
                            location = driver.find_element(By.XPATH, '//*[@id="JDCol"]/div/article/div/div[1]/div/div/div[1]/div[3]/div[1]/div[3]').text
                            #career_oppr = driver.find_element(By.XPATH, '//*[@id="JDCol"]/div/article/div/div[2]/div[1]/div[3]/div/ul/span[3]').text
                            collected_successfully = True
                            
                        except:
                            time.sleep(.5)
                    try:
                        salary_estimate = driver.find_element(By.XPATH, '//*[@id="JDCol"]/div/article/div/div[1]/div/div/div[1]/div[3]/div[1]/div[4]/span').text
                    except NoSuchElementException:
                        salary_estimate = -1
                    try:
                        company_rating = driver.find_element(By.XPATH,'//*[@id="JDCol"]/div/article/div/div[1]/div/div/div[1]/div[3]/div[1]/div[1]/div/span').text
                    except NoSuchElementException:
                        company_rating = -1
                    try:
                        career_oppr = driver.find_element(By.XPATH, '//*[@id="JDCol"]/div/article/div/div[2]/div[1]/div[4]/div/ul/span[3]').text
                    except NoSuchElementException:
                        career_oppr = -1
                    try:
                        comp_benefits = driver.find_element(By.XPATH, '//*[@id="JDCol"]/div/article/div/div[2]/div[1]/div[4]/div/ul/span[6]').text
                    except NoSuchElementException:
                        comp_benefits = -1
                    try:
                        culture_values = driver.find_element(By.XPATH, '//*[@id="JDCol"]/div/article/div/div[2]/div[1]/div[4]/div/ul/span[9]').text
                    except NoSuchElementException:
                        culture_values = -1
                    try:
                        upper_mangement = driver.find_element(By.XPATH, '//*[@id="JDCol"]/div/article/div/div[2]/div[1]/div[4]/div/ul/span[12]').text
                    except NoSuchElementException:
                        upper_mangement = -1
                    try:
                        work_life_balance = driver.find_element(By.XPATH, '//*[@id="JDCol"]/div/article/div/div[2]/div[1]/div[4]/div/ul/span[15]').text
                    except NoSuchElementException:
                        work_life_balance = -1
                        
                    try:
                        company_pros1 = driver.find_element(By.XPATH, '//*[@id="Reviews"]/div/div/div[1]/p[1]').text
                    except NoSuchElementException:
                        company_pros1 = -1
                    try:
                        company_pros2 = driver.find_element(By.XPATH, '//*[@id="Reviews"]/div/div/div[1]/p[2]').text
                    except NoSuchElementException:
                        company_pros2 = -1
                    try:
                        company_pros3 = driver.find_element(By.XPATH, '//*[@id="Reviews"]/div/div/div[1]/p[3]').text
                    except NoSuchElementException:
                        company_pros3 = -1
                    try:
                        company_pros4 = driver.find_element(By.XPATH, '//*[@id="Reviews"]/div/div/div[1]/p[4]').text
                    except NoSuchElementException:
                        company_pros4 = -1
                    try:
                        company_pros5 = driver.find_element(By.XPATH, '//*[@id="Reviews"]/div/div/div[1]/p[5]').text
                    except NoSuchElementException:
                        company_pros5 = -1


                    try:
                        company_cons1 = driver.find_element(By.XPATH, '//*[@id="Reviews"]/div/div/div[2]/p[1]').text
                    except NoSuchElementException:
                        company_cons1 = -1
                    try:
                        company_cons2 = driver.find_element(By.XPATH, '//*[@id="Reviews"]/div/div/div[2]/p[2]').text
                    except NoSuchElementException:
                        company_cons2 = -1
                    try:
                        company_cons3 = driver.find_element(By.XPATH, '//*[@id="Reviews"]/div/div/div[2]/p[3]').text
                    except NoSuchElementException:
                        company_cons3 = -1
                    try:
                        company_cons4 = driver.find_element(By.XPATH, '//*[@id="Reviews"]/div/div/div[2]/p[4]').text
                    except NoSuchElementException:
                        company_cons4 = -1
                    try:
                        company_cons5 = driver.find_element(By.XPATH, '//*[@id="Reviews"]/div/div/div[2]/p[5]').text
                    except NoSuchElementException:
                        company_cons5 = -1
                    


                    company_pros = []
                    company_cons = []
                    company_pros.extend([company_pros1, company_pros2, company_pros3, company_pros4, company_pros5])
                    company_cons.extend([company_cons1, company_cons2, company_cons3, company_cons4, company_cons5])
                    
                    jobs.append({'Job Title' : job_title,
                                'Company' : company_name,
                                'Company Rating' : company_rating,
                                'Salary' : salary_estimate,
                                'Location' : location,
                                'Career Opportunities' : career_oppr,
                                'Company Benefits' : comp_benefits,
                                'Culture & Values' : culture_values,
                                'Upper Management' : upper_mangement,
                                'Work Life Balance' : work_life_balance,
                                'Company Pros Reviews' : company_pros,
                                'Company Cons Reviews' : company_cons})
        else:
            break

    
    return pd.DataFrame(jobs)
        

```
